## Getting Started with Qiskit

The fundamental package of Qiskit is Terra that provides the basic building blocks necessary to program quantum computers.
Basic workflow is $\textbf{Build}$ and $\textbf{Execute}$


```python
import numpy as np
from qiskit import *
```

```python
# making and visualizing
qc = QuantumCircuit(3)

qc.h(0)
qc.cx(0,1)
qc.cx(0,2)
qc.draw('mpl')
```

## Simulating circuits using Qiskit Aer

```python
from qiskit import Aer

backend = Aer.get_backend('statevector_simulator')

job = backend.run(qc)
```

When we run a job, two methods are useful job.status() and job.result()

When we run a complex job on the Quantum Computer, we can check the status by that and the output by the later

```python
result= job.result()
```

Now to get the statevector of the quantum circuit

```python
outputstate = result.get_statevector(qc, decimals = 5)
print(outputstate)
```

### visualization of the result

```python
from qiskit.visualization import plot_state_city

plot_state_city(outputstate)
```

## Getting the Unitary matrix

```python
backend = Aer.get_backend('unitary_simulator')
job = backend.run(qc)
result = job.result()


print(result.get_unitary(qc, decimals = 5))
```
